### 2.3 Corregir formatos numéricos y de fecha

**🎯Objetivo:**

**Instrucciones:**

- Convierte las columnas de fecha de `traffic` a formato `datetime`. Haz el cambio a prueba de errores.
- En el dataset `eco`, limpia los valores numéricos:
    - En `city_gdp_capita`: elimina separadores de miles (`.`) y reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
    - En `unemployment_pct`: elimina el símbolo de porcentaje (`%`) y reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
    - En `population_m`: reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.

<details>
<summary>Haz clic para ver la pista</summary>
para eliminar símbolos, puedes reemplazarlos por un texto vacío.

````python
# Convertir las columnas de traffic a tipo fecha con pd.to_datetime()
traffic['update_time_utc'] = pd.to_datetime(traffic['update_time_utc'], errors="coerce",utc=True)
traffic['update_time_utc_week_ago'] = pd.to_datetime(traffic['update_time_utc_week_ago'], errors="coerce",utc=True)

# verificar el cambio
traffic.info()
````

## 🧩Paso 5: Unir movilidad y economía

Combinar datasets te permite analizar cómo se relacionan los indicadores económicos con los de movilidad.

### 5.1 Unir tráfico (tabla principal) con indicadores económicos

**🎯Objetivo:**
Combinar la información de tráfico y economía en un solo DataFrame para analizar cómo las condiciones económicas se relacionan con la movilidad urbana.

**Instrucciones**
- Selecciona solo las **columnas relevantes** de cada dataset (por ejemplo, variables clave de tráfico y de economía).
- Usa `.copy()` al crear subconjuntos para evitar modificar el dataset original.
- Une ambos DataFrames y define como **claves de unión** a `city` y `year`.
- Mantén solo las ciudades y años presentes en ambos datasets.
- Guarda el resultado en una nueva variable llamada `merged` y muestra las primeras 5 filas.


<details>
<summary>Haz clic para ver la pista</summary>
Aplica una unión de tipo "inner" para mantener las ciudades y años presentes en ambos datasets.

````python

# Seleccionar columnas clave de tráfico y economía
left_cols = ['city','country','year','jams_delay','traffic_index_live',
             'jams_length_in_kms','jams_count','mins_delay',
             'travel_time_live_per_10_kms_mins','travel_time_historic_per_10_kms_mins']

right_cols = ['city','year','city_gdp_per_capita','unemployment_percent','pm25_ug_m3','population']

# Usar .copy() para crear los dos nuevos datasets reducidos
traffic_2024_small = traffic_city_year_2024[left_cols].copy()
eco_2024_small = eco_2024[right_cols].copy()

# Unir datasets
merged = pd.merge(traffic_2024_small, eco_2024_small, on=['city', 'year'], how='inner')

# Mostrar las primeras 5 filas
merged.head(5)
````

### Comentarios sobre el BoxPlot de Jams_Delay
En general, se aprecia que la mayoría de los datos se colocan por debajo del promedio general (629.52). Incluso hay muy pocos valores dentro del primer cuartil. La mediana también es muy baja y se halla mucho más cerca al primer cuartil que al tercero o incluso al promedio. La gran mayoría de los datos cayeron dentro del segundo y tercer cuartil, pero pienso que este BoxPlot está inflado por el outlier que se aprecia fuera del área y que además el valor máximo también es muy elevado y aparace lejano a donde están la mayoría de los datos. Sería importante realizar una nueva medición, eliminando este outlier revisando el valor máximo registrado, ya que están distorsionando la representación del gráfico y su interpretación.


````python
# En esta celda busqué los outliers, es decir, aquellos registros que están muy lejos de la media. En efecto, puedo observar que solamente hay cuatro registros que están por encima de los 900 en Jams_Delay
# y el más extremo de ellos (el outlier) tiene un valor de 2,833, disparando el BoxPlot anterior
valores_altos = merged[merged["jams_delay"]>900].sort_values(by="jams_delay",ascending=False)
valores_altos
````

````python
# Crear histograma para ver la distribución de la economía (city_gdp_capita)

merged["city_gdp_per_capita"].hist(bins=5,figsize=(10,5))
plt.title("Distribución del PIB per cápita (2024)")
plt.xlabel("PIB per cápita (USD)")
plt.ylabel("Número de ciudades")
plt.show()
````

### Comentarios sobre el histograma
La mayoria de las ciudades tiene un PIB per cápita que va entre los 7,000 y 12,000, lo cual podemos ver en los picos más altos del histograma. El gráfico nos muestra también que solamente una ciudad tiene valores por debajo de los 7,000 y otra más por encima de los 20,000 USD

````python
# Ciudades que están en la parte más alta del histograma (PIB entre 7,500 y 16,000 USD).
# Se puede ver que son prácticamente puras ciudades brasileñas. 
ciudades_PIB_promedio = merged[merged["city_gdp_per_capita"].between(7500,16000)]
ciudades_PIB_promedio
````

````python
#Debido a que los valores entre "jams_delay" y "city_gdp_per_capita" son muy distintos, aplicaré una normalización de los valores para que se puedan comparar en la gráfica de columnas
merged['jams_delay_norm'] = (merged['jams_delay'] - merged['jams_delay'].min()) / (merged['jams_delay'].max() - merged['jams_delay'].min())
merged['gdp_norm'] = (merged['city_gdp_per_capita'] - merged['city_gdp_per_capita'].min()) / (merged['city_gdp_per_capita'].max() - merged['city_gdp_per_capita'].min())

# Gráfico de barras para comparar jams_delay y city_gdp_capita por ciudad
merged.plot(kind="bar",x="city", y=['jams_delay_norm', 'gdp_norm'])
plt.xticks(rotation=90)
plt.show()
````
